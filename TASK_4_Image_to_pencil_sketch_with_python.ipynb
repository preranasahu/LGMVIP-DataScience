```python
from google.colab.patches import cv2_imshow
```

```python
import cv2
```

```python

img= cv2.imread('/content/WhatsApp Image 2021-09-06 at 22.12.46.jpeg')
cv2_imshow(img)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(img_gray)
```

```python
img_invert = cv2.bitwise_not(img_gray)
```

```python
cv2_imshow(img_invert)
```

```python
blurred = cv2.GaussianBlur(img_invert, (21, 21), 0)
cv2_imshow(blurred)
```

```python
final_blur = 255 - blurred
pencil = cv2.divide(img_gray, final_blur, scale=256.0)
cv2_imshow(pencil)
```

```python
cv2_imshow(img)
cv2_imshow( pencil)
```
